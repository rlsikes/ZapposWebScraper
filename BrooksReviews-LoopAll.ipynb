```python

def __get_Reviews__(soup): 
    review_data = list()
    reviews = soup.find_all('p', class_='reviewSummary')
    if not len(reviews) ==0:
        for i in range(0,len(reviews)):
            review_data.append(reviews[i].get_text())
        return(review_data)
    else: 
        review_data.append(" ")
        print("reviewSummary is blank")
        return(review_data)
```
